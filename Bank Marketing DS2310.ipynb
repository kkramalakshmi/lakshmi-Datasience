# Model Buiding

```python
Y= df1['subscribed']
X= df1.drop('subscribed', axis=1)
```

```python
X
```

```python
Y
```

```python
#generating dummy values on the dataset
X= pd.get_dummies(X)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X 
```

we have scaled data using standard scalarisation to overcome the issue of biasseness because feature columns are of float data type so  my model biased. 

```python
maxAccu=0
maxRS=0 
for i in range(1,50):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=i)
    RFR=RandomForestClassifier()
    RFR.fit(X_train,Y_train)
    pred=RFR.predict(X_test)
    acc=accuracy_score(Y_test,pred) 
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is-",maxAccu,"at randomstate is -",maxRS)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
RFC=RandomForestClassifier()
RFC.fit(X_train,Y_train)
predRFC=RFC.predict( X_test)
print(accuracy_score(Y_test,predRFC))
print(confusion_matrix(Y_test,predRFC))
print(classification_report(Y_test,predRFC))
```

```python
LR=LogisticRegression()
LR.fit(X_train,Y_train)
predLR=LR.predict( X_test)
print(accuracy_score(Y_test,predLR))
print(confusion_matrix(Y_test,predLR))
print(classification_report(Y_test,predLR))
```

```python
# check accuracy for support vector machine classifier
svc=SVC()
svc.fit(X_train,Y_train)
predsvc=svc.predict( X_test)
print(accuracy_score(Y_test,predsvc))
print(confusion_matrix(Y_test,predsvc))
print(classification_report(Y_test,predsvc))
```

```python
GB=GradientBoostingClassifier()
GB.fit(X_train,Y_train)
predGB=GB.predict( X_test)
print(accuracy_score(Y_test,predGB))
print(confusion_matrix(Y_test,predGB))
print(classification_report(Y_test,predGB))
```

```python
abc=AdaBoostClassifier()
abc.fit(X_train,Y_train)
predabc=abc.predict( X_test)
print(accuracy_score(Y_test,predabc))
print(confusion_matrix(Y_test,predabc))
print(classification_report(Y_test,predabc))
```

```python
bc=BaggingClassifier()
bc.fit(X_train,Y_train)
predbc=bc.predict( X_test)
print(accuracy_score(Y_test,predbc))
print(confusion_matrix(Y_test,predbc))
print(classification_report(Y_test,predbc))
```

```python
et=ExtraTreesClassifier()
et.fit(X_train,Y_train)
predet=et.predict( X_test)
print(accuracy_score(Y_test,predet))
print(confusion_matrix(Y_test,predet))
print(classification_report(Y_test,predet))
```

```python
#checking cv for random forest classifier
score=cross_val_score(RFC,X,Y)
print(score)
print(score.mean())
print("Differences between Acuracy and cv score is-",accuracy_score(Y_test,predRFC)-score.mean())
```

```python
#checking cv for LogisticRegression
score=cross_val_score(LR,X,Y)
print(score)
print(score.mean())
print("Differences between Acuracy and cv score is-",accuracy_score(Y_test,predLR)-score.mean())
```

```python
#checking cv for SVC
score=cross_val_score(svc,X,Y)
print(score)
print(score.mean())
print("Differences between Acuracy and cv score is-",accuracy_score(Y_test,predsvc)-score.mean())
```

```python
#checking cv for GradientBoostingClassifier
score=cross_val_score(GB,X,Y)
print(score)
print(score.mean())
print("Differences between Acuracy and cv score is-",accuracy_score(Y_test,predGB)-score.mean())
```

```python
#checking cv for AdaBoostClassifier
score=cross_val_score(abc,X,Y)
print(score)
print(score.mean())
print("Differences between Acuracy and cv score is-",accuracy_score(Y_test,predabc)-score.mean())
```

```python
#checking cv for BaggingClassifier
score=cross_val_score(bc,X,Y)
print(score)
print(score.mean())
print("Differences between Acuracy and cv score is-",accuracy_score(Y_test,predbc)-score.mean())
```

```python
#checking cv for ExtraTreeClassifier
score=cross_val_score(et,X,Y)
print(score)
print(score.mean())
print("Differences between Acuracy and cv score is-",accuracy_score(Y_test,predet)-score.mean())
```

BaggingClassifier is our best model as the differences between accuracy and cv score is least

BaggingClassifier is our Best Model

Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'estimator':[None],
            'n_estimators':[10],
            'max_samples':[1.0], 
            'max_features':[1.0], 
            'bootstrap':[True], 
            'bootstrap_features':[False],
            'oob_score':[False], 
            'warm_start':[False], 
            'n_jobs':[None], 
            'random_state':[None], 
            'verbose':[0]}
```

```python
GCV=GridSearchCV(BaggingClassifier(),parameters,cv=5)
```

```python
GCV.fit(X_train,Y_train)
```

```python
GCV.best_params_
```

```python
Final_model=BaggingClassifier(bootstrap=True,
 bootstrap_features=False,
 estimator=None,
 max_features=1.0,
 max_samples=1.0,
 n_estimators=10,
 n_jobs=None,
 oob_score=False,
 random_state=None,
 verbose=0,
 warm_start=False)
Final_model.fit(X_train,Y_train)
pred=Final_model.predict(X_test)
acc=accuracy_score(Y_test,pred)
print(acc*100)
```

```python
from sklearn import metrics
fpr,tpr,thresholds=metrics.roc_curve(Y_test,pred)
roc_auc=metrics.auc(fpr,tpr)
display=metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=Final_model)
display.plot()
```
