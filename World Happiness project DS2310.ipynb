```python
Index=0
labels = df['Happiness Score']
features = df.drop('Happiness Score', axis=1)

for col in features.items():
    plt.figure(figsize=(2,3))
    sns.barplot(x=labels, y=col[Index], data=df, palette='husl')
plt.tight_layout()
plt.show()
```

```python
Index=0
labels = df['Happiness Score']
features = df.drop('Happiness Score', axis=1)

for col in features.items():
    plt.figure(figsize=(2,3))
    sns.scatterplot(x=labels, y=col[Index], data=df, color="red")
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
plt.figure()
sns.heatmap(df.corr(),annot=True)
plt.xlabel('Figure')
plt.ylabel('Features_Name')
plt.title('Discriptive_Graph')
plt.show()
```

Above  graph are showing us how is the correlation between each variable, we will concentrate the highly correlated variables. Happiness Rank and Happiness Score have negative correlation, while Happiness Score is increasing, Ranking is going to decrease (1 is the top, 158 is the last ranking). Therefore, we will analyze Happiness Score relations with GDP, Life Expectations, Freedom and Trust to Government Corruption as these values are highly correlated to each other. GDP is the main factor which is effecting others as Family, Life Expectations and Freedom.
