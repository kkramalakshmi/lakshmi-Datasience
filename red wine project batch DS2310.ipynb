```python
df = df.drop('free sulfur dioxide', axis=1)
df
```

```python
from scipy.stats import zscore
```

```python
out_features=df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','total sulfur dioxide','density','pH','sulphates','alcohol']]
z=np.abs(zscore(out_features))
z
```

```python
np.where(z>3)
```

```python
z.iloc[15,6]
```

```python
#now removing the data above zscore and creating new df 
df1=df[(z<3).all(axis=1)]
df1.shape
```

```python
print('old dataframe',df.shape[0])
print('New dataframe',df1.shape[0])
```

```python
print('Data loss percentage-',((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
df1
```

it is affordable loss of data now will check with IQR method

```python
Q1=out_features.quantile(0.25)
Q3=out_features.quantile(0.75)
IQR=Q3-Q1
df2=df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
```

```python
df2.shape
```

```python
print('Data loss percentage after removing outliers with IQR method-',((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# feature scaling

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X 
```

we have scaled data using standard scalarisation to overcome the issue of biasseness because feature columns are of float data type so  my model biased. 

# finding best random state

```python
maxAccu=0
maxRS=0
for i in range(1,250):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=i)
    RFR=RandomForestClassifier()
    RFR.fit(X_train,Y_train)
    pred=RFR.predict(X_test)
    acc=accuracy_score(Y_test,pred) 
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is-",maxAccu,"at randomstate is -",maxRS)
```

# creating train test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=maxRS)
```

# classification Algoritham

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
RFC=RandomForestClassifier()
RFC.fit(X_train,Y_train)
predRFC=RFC.predict( X_test)
print(accuracy_score(Y_test,predRFC))
print(confusion_matrix(Y_test,predRFC))
print(classification_report(Y_test,predRFC))
```

accuracy of this model is 97.1%

```python
LR=LogisticRegression()
LR.fit(X_train,Y_train)
predLR=LR.predict( X_test)
print(accuracy_score(Y_test,predLR))
print(confusion_matrix(Y_test,predLR))
print(classification_report(Y_test,predLR))
```

accuracy of this model is 93%

```python
# check accuracy for support vector machine classifier
svc=SVC()
svc.fit(X_train,Y_train)
predsvc=svc.predict( X_test)
print(accuracy_score(Y_test,predsvc))
print(confusion_matrix(Y_test,predsvc))
print(classification_report(Y_test,predsvc))
```

accuracy of this model is 94%

```python
GB=GradientBoostingClassifier()
GB.fit(X_train,Y_train)
predGB=GB.predict( X_test)
print(accuracy_score(Y_test,predGB))
print(confusion_matrix(Y_test,predGB))
print(classification_report(Y_test,predGB))
```

accuracy of this model is 95%

```python
abc=AdaBoostClassifier()
abc.fit(X_train,Y_train)
predabc=abc.predict( X_test)
print(accuracy_score(Y_test,predabc))
print(confusion_matrix(Y_test,predabc))
print(classification_report(Y_test,predabc))
```

accuracy of this model is 93%
